```python
# 1 Построение простого графика
# Напишите код, который строит график функции y = x^2
# на интервале [−10, 10].
# Добавьте заголовок, подписи осей и сетку.

import matplotlib.pyplot as plt
import numpy as np

x = np.linspace(-10, 10, 100)

y = x**2

plt.figure(figsize=(8, 4))

plt.plot(x, y)
plt.title("График функции y = x^2")
plt.xlabel("Ось X")
plt.ylabel("Ось Y")
plt.grid()

plt.show()
```

```python
# 2 Настройка стилей и цветов
# y = x (синяя, пунктирная линия),
# y = x^2 (зеленая, штрих-пунктирная линия),
# y = x^3 (красная, сплошная линия).
# Добавьте легенду и сделайте оси одинакового масштаба.

import matplotlib.pyplot as plt
import numpy as np

x = np.linspace(-2, 2, 10)

y1 = x
y2 = x**2
y3 = x**3

plt.figure(figsize=(8, 4))

g1 = plt.plot(x, y1)
plt.setp(g1, linestyle='--', color='blue', label='y = x')
g2 = plt.plot(x, y2)
plt.setp(g2, linestyle='-.', color='green', label='y = x^2')
g3 = plt.plot(x, y3)
plt.setp(g3, linestyle='-', color='red', label='y = x^3')

plt.title("Графики функций y = x, y = x^2, y = x^3")
plt.xlabel("Ось X")
plt.ylabel("Ось Y")
plt.grid()
plt.legend()
plt.axis('equal')

plt.show()
```

```python
# 4 Гистограмма распределения
# Сгенерируйте 1000 случайных чисел из нормального распределения с
# параметрами μ = 0, σ = 1 и постройте их гистограмму с 30 бинами. Добавьте
# вертикальную линию в среднем значении.

import matplotlib.pyplot as plt
import numpy as np

mo = 0
otkl = 1
n = 1000
data = np.random.normal(mo, otkl, n)

plt.figure(figsize=(8, 6))
num_bins = 30
plt.figure(figsize=(8, 6))
plt.hist(data, bins=num_bins, facecolor='black')

mean = np.mean(data)
plt.axvline(mean, color='red', linewidth=1)

plt.show()
```

```python
# 7 Трехмерный график поверхности
# Используя mpl_toolkits.mplot3d , постройте 3D-график функции z = sin(√x^2 + y
# ^2)
# на сетке значений x, y в диапазоне [−5, 5].

import matplotlib.pyplot as plt
import numpy as np
from mpl_toolkits.mplot3d import Axes3D

x = np.linspace(-5, 5, 50)
y = np.linspace(-5, 5, 50)
x, y = np.meshgrid(x, y)
z = np.sin(np.sqrt(x**2 + y**2))

fig = plt.figure(figsize=(8, 6))
ax = fig.add_subplot(111, projection='3d')

surf = ax.plot_surface(x, y, z, cmap='inferno')

ax.set_xlabel('x')
ax.set_ylabel('y')
ax.set_zlabel('z')
ax.set_title('График z = sin(sqrt(x^2 + y^2))')

plt.show()
```

```python
# 8 Множественные подграфики (subplots)
#Постройте четыре графика в одной фигуре (2 × 2 сетка):


import matplotlib.pyplot as plt
import numpy as np

x = np.linspace(0, 4 * np.pi, 100)

y1 = np.sin(x)
y2 = np.sin(2 * x)
y3 = np.sin(x + np.pi / 4)
y4 = np.sin(x - np.pi / 4)


fig, axs = plt.subplots(2, 2, figsize=(8, 6))

axs[0, 0].plot(x, y1, color='blue')
axs[0, 0].set_title('sin(x)')
axs[0, 0].set_xlabel('x')
axs[0, 0].set_ylabel('y')
axs[0, 0].grid()

axs[0, 1].plot(x, y2, color='red')
axs[0, 1].set_title('sin(2x)')
axs[0, 1].set_xlabel('x')
axs[0, 1].set_ylabel('y')
axs[0, 1].grid()

axs[1, 0].plot(x, y3, color='green')
axs[1, 0].set_title('sin(x + pi/4)')
axs[1, 0].set_xlabel('x')
axs[1, 0].set_ylabel('y')
axs[1, 0].grid()

axs[1, 1].plot(x, y4, color='purple')
axs[1, 1].set_title('sin(x - pi/4)')
axs[1, 1].set_xlabel('x')
axs[1, 1].set_ylabel('y')
axs[1, 1].grid()

plt.tight_layout()

plt.show()
```

```python
# 3 Индивидуальное Вариант 6
# Вычислите определенный интеграл для функции:
# f(x) = x^3 − 2x^2 + x
# на отрезке [−1, 2].

import numpy as np
import matplotlib.pyplot as plt
from scipy import integrate

def f(x):
    return x**3 - 2*x**2 + x

a = -1
b = 2

x = np.linspace(a - 0.5, b + 0.5, 400)
y = f(x)

result, error = integrate.quad(f, a, b)

fig, ax = plt.subplots(figsize=(8, 6))

ax.plot(x, y, label='f(x) = x^3 - 2x^2 + x', color='blue')
x_fill = np.linspace(a, b, 400)
y_fill = f(x_fill)
ax.fill_between(x_fill, y_fill, where=(y_fill > 0), color='green')
ax.fill_between(x_fill, y_fill, where=(y_fill < 0), color='red')

ax.axhline(0, color='black')
ax.axvline(a, color='black', linestyle='--')
ax.axvline(b, color='black', linestyle='--')

ax.set_title('Определенный интеграл функции')
ax.set_xlabel('x', fontsize=12)
ax.set_ylabel('f(x)', fontsize=12)

ax.grid()

ax.text(0.5, 0.8, f'Значение интеграла: {result:.4f}', fontsize=12,
        bbox=dict(facecolor='white', alpha=0.8))

plt.show()
```

```python
# 4 Индивидуальное Вариант 6
# Постройте поверхность:
# f(x, y) = 1 / (1 + x**2 + y**2)
# на x, y ∈ [−3, 3]

import numpy as np
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D

def f(x, y):
    return 1 / (1 + x**2 + y**2)

x_min, x_max = -3, 3
y_min, y_max = -3, 3

x = np.linspace(x_min, x_max, 100)
y = np.linspace(y_min, y_max, 100)
X, Y = np.meshgrid(x, y)

Z = f(X, Y)


fig = plt.figure(figsize=(8, 6))
ax = fig.add_subplot(111, projection='3d')
surf = ax.plot_surface(X, Y, Z, cmap='inferno')


ax.set_xlabel('x', fontsize=12)
ax.set_ylabel('y', fontsize=12)
ax.set_zlabel('f(x, y)', fontsize=12)
ax.set_title('Поверхность', fontsize=14)

plt.show()
```
